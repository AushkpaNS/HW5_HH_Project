1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
#ваш код здесь
# Создаём колонку 'Образование'
# при помощи метода apply и lambda-функции определяем категорию образования,
# для этого проверям с каких слов начинается строка в колонке 'Образование и ВУЗ'
# если строка в колонке 'Образование и ВУЗ' не начинается с ожидаемых слов, то в колонке 'Образование' будет пустое значение
df['Образование'] = df['Образование и ВУЗ'].apply(lambda x: 
    'высшее' if x.upper().startswith('ВЫСШЕЕ') else 
    'неоконченное высшее' if x.upper().startswith('НЕОКОНЧЕННОЕ ВЫСШЕЕ') else 
    'среднее специальное' if x.upper().startswith('СРЕДНЕЕ СПЕЦИАЛЬНОЕ') else 
    'среднее' if x.upper().startswith('СРЕДНЕЕ') else 
    None
)

print('Количество строк с неопределённой категорией в новом столбце "Образование":')
print(df['Образование'].isnull().sum()) # проверяем, что у всех строк определилась одна из ожидаемых категорий. Ожидаемый резульат: 0
print()

df['Образование'] = df['Образование'].astype('category') # приводим к категориальному типу данных

df = df.drop(columns=['Образование и ВУЗ']) # удаляем колонку 'Образование и ВУЗ'

print('Количество строк в каждой категории "Образование":')
print(df['Образование'].value_counts())
```

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '.

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
def experience_in_months(expStr):
    # Проверяем на скрытые и обычные пропуски, на всякий случай применяем strip, который удаляет пробелы в начале и конце строки
    if pd.isna(expStr) or expStr.strip().upper() == "НЕ УКАЗАНО":
        total_months = np.nan
    else:
        years = 0
        months = 0
        exit_flg = False
        year_flg = False # признак того, что года уже считали
        words = expStr.split() # рзбиваем строку на слова
        for i in range(len(words)): # перебираем слова
            if not year_flg and words[i].lower() in ['год', 'года', 'лет']: # если встретили слово год, года, лет и года еще не считали
                if i > 0 and words[i-1].isdigit(): # проверяем, что предшествующее слово является числом
                    years = int(words[i-1]) # запишем в years
                else:
                    print('Предупреждение: встретилось слово год/года/лет, но перед ним нет числа!')
                year_flg = True
            else:
                if words[i].lower() in ['месяц', 'месяца', 'месяцев']: # если встретили слово месяц,'месяца, месяцев
                    if i > 0 and words[i-1].isdigit():  # проверяем, что предшествующее слово является числом
                        months = int(words[i-1])
                    else:
                        print('Предупреждение: встретилось слово месяц/месяца/месяцев, но перед ним нет числа!')
                    exit_flg = True # если считали месяцы, значит можно остановиться
                elif year_flg and not words[i].isdigit(): # если считали года, но после этого не число и не месяцы, то можно остановиться
                    exit_flg = True

            if exit_flg:
                break
        total_months = years * 12 + months
        if total_months == 0:
            print('Предупреждение: стаж равен 0 месяцев!')
    return total_months

# Применяем функцию к столбцу 'Опыт работы' и создаём новый столбец 'Опыт работы (месяц)'
df['Опыт работы (месяц)'] = df['Опыт работы'].apply(experience_in_months)

df = df.drop(columns=['Опыт работы']) # удаляем колонку 'Опыт работы'

print('Медианный опыт работы:')
print(df['Опыт работы (месяц)'].median())
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

    Список городов-миллионников:
    
   <code>million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']

    Инфорация о метро, рядом с которым проживает соискатель нас не интересует.
* Признак **"Готовность к переезду"** должен иметь два возможных варианта: True или False. Обратите внимание, что возможны несколько вариантов описания готовности к переезду в признаке "Город, переезд, командировки". Например:

    * … , готов к командировкам , … 
    * … , готова к редким командировкам , …
    * … , не готов к командировкам , …
    
    Нас интересует только сам факт готовности к командировке.
    
    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
# Список городов-миллионников в нижнем регистре, так далее сравнение производится в нижнем регистре
million_cities = [
    'новосибирск', 'екатеринбург', 'нижний новгород', 'казань', 
    'нелябинск', 'омск', 'самара', 'ростов-на-дону', 'уфа', 
    'красноярск', 'пермь', 'воронеж', 'волгоград'
]

# Функция для преобразования данных
def get_info(exp_str):
    if pd.isna(exp_str):
        return np.nan, np.nan, np.nan
    
    exp_str = exp_str.lower() # приводим к нижнему регистру для сравнения
    exp_str = exp_str.replace(' ', '') # удаляем все пробелы

    parts = exp_str.split(',') # разбиваем строку на блоки информации с разделителем ','

    # Определеняем город
    city_info = parts[0]
    if city_info == 'москва':
        city = 'Москва'
    elif city_info == 'санкт-петербург':
        city = 'Санкт-Петербург'
    elif city_info in million_cities:
        city = 'город-миллионник'
    else:
        city = 'другие'

    # Определяем готовность к переезду
    ready_to_relocate = any(keyword in exp_str for keyword in [',готовкпереезду', ',готовакпереезду', ',хочупереехать'])

    # Определяем готовность к командировкам
    ready_for_business_trip = any(keyword in exp_str for keyword in [',готовккомандировкам', ',готовкредкимкомандировкам', ',готоваккомандировкам', ',готовакредкимкомандировкам'])

    return city, ready_to_relocate, ready_for_business_trip

# Применяем функцию и создаем новые колонки
df[['Город', 'Готовность к переезду', 'Готовность к командировкам']] = df['Город, переезд, командировки'].apply(get_info).apply(pd.Series)

# Преобразуем колонки 'Готовность к переезду' и 'Готовность к командировкам' к типу bool
df['Готовность к переезду'] = df['Готовность к переезду'].astype(bool)
df['Готовность к командировкам'] = df['Готовность к командировкам'].astype(bool)

# Удаляем столбец 'Город, переезд, командировки'
df = df.drop(columns=['Город, переезд, командировки'])


# Считаем процент соискателей, живущих в Санкт-Петербурге
total_candidates = df.shape[0]
spb_candidates = df[df['Город'] == 'Санкт-Петербург'].shape[0]
percentage_spb = round((spb_candidates / total_candidates) * 100)
print(f"Процент соискателей, живущих в Санкт-Петербурге: {percentage_spb}%")

# Считаем процент соискателей, готовых одновременно к переездам и командировкам
ready_for_both = df[(df['Готовность к переезду']) & (df['Готовность к командировкам'])].shape[0]
percentage_ready_for_both = round((ready_for_both / total_candidates) * 100)
print(f"Процент соискателей, готовых одновременно к переездам и командировкам: {percentage_ready_for_both}%")
```

Создайте новый DataFrame из полученного файла. В полученной таблице нас будут интересовать столбцы:

Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь
# функция возвращает число и валюту по строке столбца 'ЗП'
def getZp(row):
    parts = row.split(' ', 1)  # Разделяем строку по первому пробелу
    value = int(parts[0])  # Первое значение - это число
    if len(parts) > 1:
        currency = parts[1].strip() # Остальная часть - это валюта без лишних пробелов
    else:
        currency = ''
    return value, currency

# функция по числу, валюте, пропорции и курсу возвращает сумму в рублях
def getZpRub(value, currency, proportion, close):
    if currency == 'руб.':
        zp = float(value)
    else:
        zp = round(float(value) * float(close) / float(proportion), 2)
    return zp

exchange_rates_df = pd.read_csv("data/ExchangeRates.csv", encoding='utf-8', sep=',') #считываем таблицу курсов валют

df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме'], format='%d.%m.%Y %H:%M') # приводим признак 'Обновление резюме' к типу datetime
exchange_rates_df['date'] = pd.to_datetime(exchange_rates_df['date'], format='%d/%m/%y') # приводим признак 'Обновление резюме' к типу datetime
df['Обновление резюме'] = df['Обновление резюме'].dt.date # берём только дату
exchange_rates_df['date'] = exchange_rates_df['date'].dt.date # берём только дату

# Применяем функцию getZp и создаем новые колонки: 'ЗП число' и 'ЗП валюта'
df[['ЗП число', 'ЗП валюта']] = df['ЗП'].apply(getZp).apply(pd.Series)

print('Уникальные валюты в исходных данных df: ')
print(df['ЗП валюта'].unique())
print()
print('Уникальные валюты в таблице курсов валют ExchangeRates.csv: ')
print(exchange_rates_df['currency'].unique())

# создаём словарь соответствия валют из исходых данных в df и из таблицы курсов валют
currency_dict = {'бел.руб.': 'BYN',
                 'KZT': 'KZT',
                 'EUR': 'EUR',
                 'USD': 'USD', 
                 'грн.':  'UAH',
                 'сум': 'UZS',
                 'KGS': 'KGS',
                 'AZN': 'AZN'}

df['ЗП валюта'] = df['ЗП валюта'].replace(currency_dict) # заменям значения в колонке 'ЗП валюта' в соответствии со словарем currency_dict

# присоединям интересующие нас колонки таблицы курсов валют к df 
df = df.merge(exchange_rates_df[['currency', 'date', 'proportion', 'close']],
                  left_on=['ЗП валюта', 'Обновление резюме'],
                  right_on=['currency', 'date'],
                  how='left')

# проверяем, что количество строк не изменилось. Ожидаемое значение: 44744
print(df.shape[0]) 
# Проверяем, что для всех строк из df, где валюта не равна 'руб.', определился курс валют.
mask = df['ЗП валюта'] != 'руб.'
# Проверяем, есть ли пустые значения в колонке 'close'
if df[mask]['close'].isnull().any():
    print("Есть строки с пустыми значениями курса валют.")
    print()
else:
    print("Все строки с 'ЗП валюта' != 'руб.' имеют курс валюты.")
    print()

# Создаем новую колонку 'ЗП (руб)'
df['ЗП (руб)'] = df.apply(lambda x: getZpRub(x['ЗП число'], x['ЗП валюта'], x['proportion'], x['close']), axis=1)

# удаляем все лишние колонки, которые добавили в процессе и исходную колонку 'ЗП'
df = df.drop(columns=['ЗП'])
df = df.drop(columns=['ЗП число'])
df = df.drop(columns=['ЗП валюта'])
df = df.drop(columns=['currency'])
df = df.drop(columns=['date'])
df = df.drop(columns=['proportion'])
df = df.drop(columns=['close'])

print('Медианная желаемая зарплата в рублях:')
print(df['ЗП (руб)'].median())
```

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# ваш код здесь
def outliers_z_score_mod(data, feature, left = 3, right = 3):
    x = np.log(data[feature])
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    # Выводим коэффициент ассиметрии
    print(x.skew())
    # Построение гистограммы
    plt.figure(figsize=(12, 6))
    histplot = sns.histplot(x, bins=20, kde=True, log_scale=True)
    mean_log_age = x.mean()
    histplot.axvline(mean_log_age, color='k', lw=2, label='Среднее')
    histplot.axvline(lower_bound, color='r', lw=2, linestyle='--', label=f'нижняя граница {left} сигмы')
    histplot.axvline(upper_bound, color='g', lw=2, linestyle='--', label=f'верхняя граница {right} сигмы')
    plt.legend()
    plt.title(f'Распределение {feature} в логарифмическом масштабе')
    plt.xlabel(feature)
    plt.ylabel('Частота')
    plt.show()
    return outliers, cleaned

outliers, cleaned = outliers_z_score_mod(df, 'Возраст', left = 3, right = 4)
df = cleaned
```

ваш коммментарий здесь  
 - Логарифмическое распределение имеет правостороннюю ассиметрию. Об этом говорит положительный коэффициент ассимметрии и это можно разглядеть на графике (правый "хвост" более вытянут).
